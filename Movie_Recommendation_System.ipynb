**Recommender System**
is a type of software application or algorithm designed to provide personalized suggestions or recommendations to users. The goal of a recommender system is to help users discover items (such as products, services, information, or content) that they might be interested in, based on their preferences and behavior.

**Types of Recommender System**


1- **Content Based** filtering recommends items to users based on the attributes and features of the items themselves, as well as the user's preferences. It involves analyzing the content of items and building a profile for each user based on their historical preferences.

2 - **Collaborative**  filtering methods make recommendations based on the preferences and behavior of similar users. If User A has similar preferences to User B, and User B likes a particular item, the system may recommend that item to User A.

As it has multiple dictionary in an array<br>
We want it in this format<br>
["Action","Adventure","Fantasy","SciFi"]<br><br>
Creating a function to do this

```python
import ast
```

ast.literal_eval converts the passed parameter(string) into list

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
df["genres"] = df["genres"].apply(convert)
```

We will gonna take only three lead actors and make a function for it.

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
           L.append(i["name"])
           counter = counter + 1
        else:
          break

    return L
```

```python
df["cast"] = df["cast"].apply(convert3)
```

We want only those dictionary in which "job" key has "Director" value<br>
for this let's create a function

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director" :
           L.append(i["name"])
           break
    return L
```

```python
df["crew"] = df["crew"].apply(fetch_director)
```

there are many repeated words in tag columns like action , actions,activity,activities
So for removing this we will use Stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

We are creating a function for stemming

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

applying this function in tag column of  our dataset

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

**Text Vectorization**<br><br>
It is the process of converting text data into numerical vectors or arrays that can be used as input for machine learning models, natural language processing (NLP) tasks, and other data analysis techniques. Since most machine learning algorithms and models work with numerical data, text vectorization is a crucial step when dealing with text-based information.

We will be using **Bag of Words (BoW)**: In this approach, a document is represented as an unordered set of words, ignoring grammar and word order but keeping track of word frequency. Each unique word in the text is assigned a unique index, and the vector for a document is created by counting the occurrences of each word.

We will be removing stopwords like i, to etc. from tag column and make tags as vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = "english")
```

Applying countvectorizer to tags and getting an numpy array

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

let's check shape of numpy array

```python
vectors.shape
```

let's check the most repeated or common 5000 words

```python
cv.get_feature_names_out()
```

As every movie is a vector now then now<br>
We will calculate cosine distance between every movie with every movie<br><br>
As cosine distance is greater ,movies will have less similarities


```python
from sklearn.metrics.pairwise import cosine_similarity
```

we will store the cosine distance in a similarity matrix

```python
similarity = cosine_similarity(vectors)
```

let's check shape of similarity matrix

```python
similarity.shape
```

Let's make a recommend function which will return me the movie which has very less cosine distance from the given movie

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

let's test our function

```python
recommend("Avatar")
```
